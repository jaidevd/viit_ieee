### Exercise 1: Create a simple sine wave.

```python
x = linspace(-2*pi, 2*pi, 1000)
```

```python
y = sin(x)
```

```python
figsize(10,10)
plot(x, y)
xlabel("x")
ylabel("sin(x)")
grid()
title("Sine Wave")
```

### Exercise 2: Creating Gaussian signals

```python
mu = 0
sigma = 0.1
f = random.normal(loc=mu, scale=sigma, size=(10000,))
figsize(10,10)
plot(f)
```

```python
h = hist(f, bins=100)
```

```python
sigmean = f.mean()
sigstd = f.std()
sigmin = f.min()
sigmax = f.max()
```

```python
axis([-0.5, 0.5, 0, 350])
h = hist(f, bins=100)
vlines(sigmean, 0, 350, color='g', linewidth=3, label="mean")
vlines(sigmean + sigstd, 0, 350, color="r", linewidth=3, label="std")
vlines(sigmean - sigstd, 0, 350, color="r", linewidth=3)
vlines(sigmin, 0, 350, color="c", linewidth=3, label="min")
vlines(sigmax, 0, 350, color="c", linewidth=3, label="max")
legend()
show()
```
