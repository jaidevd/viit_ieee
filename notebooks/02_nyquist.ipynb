# The Sampling Theorem

> A signal with a bandwidth $W$ can be reconstructed faithfully from its discrete samples if and only if the signal has been sampled with a $F_{s}$ such that:
> $F_{s} \geq 2W$

```python
def create_sine(freq, t0, t1, size):
    """
    Create a sine wave starting at time `t0`, ending at time `t1`, of length `size` and having a frequency of `freq`.
    """
    t = linspace(t0, t1, size)
    return t, sin(2*pi*freq*t)
```

```python
def sample_signal(f, t, fSample):
    """
    Sample a signal `f`, with time characteristics `t` at a sampling rate `fSample`
    """
    tStart = t[0]
    tStop = t[len(t)-1]
    duration = tStop - tStart
    nSamples = duration * fSample
    tStart = t[0]
    tStop = t[len(t)-1]
    duration = tStop - tStart
    nSamples = duration * fSample
    inds = linspace(0, len(f)-1, nSamples).astype(int)
    return t[inds], f[inds]
```

```python
# Create a signal with frequency 100 Hz
```

```python
t, f = create_sine(50, 0, 1, 1000)
figsize(10,10)
plot(t,f)
```

```python
sample_freqs = [5, 10, 50, 100, 150]
```

```python
from scipy.interpolate import splrep, splev
```

```python
for i in range(5):
    fs = sample_freqs[i]
    ts, fSampled = sample_signal(f, t, fs)
    interpolator = splrep(ts, fSampled)
    interpolated = splev(t, interpolator)
    subplot(5, 1, i+1)
    plot(t, f, 'b', ts, fSampled, 'r.', t, interpolated,'g')
```

### Fourier transform and the Sampling Theorem

```python
for i in range(5):
    fs = sample_freqs[i]
    ts, fSampled = sample_signal(f, t, fs)
    interpolator = splrep(ts, fSampled)
    interpolated = splev(t, interpolator)
    subplot(5, 1, i+1)
    ffx = fft.fftfreq(len(interpolated))
    ffx = fft.fftshift(ffx)
    ff = fft.fft(interpolated, axis=0)
    ff = fft.fftshift(ff)
    plot(ffx[len(ffx)/2:], abs(ff)[len(ff)/2:])
    grid()
    
```
