# Convolution and Windowing

```python
# Create a Gaussian signal and smooth it using Convolution
y = random.normal(0, 0.1, size=(1024,))
window = ones((8,))/8
Y = convolve(y, window)
plot(arange(1024), y, 'r', label='original')
plot(arange(len(Y)), Y, 'g', label='smoothed')
legend()
grid()
```

# Using prebuilt windows in Python

```python
# Create some windows
from scipy.signal import (hamming, hanning, kaiser, nuttall, blackman, blackmanharris, bartlett, barthann, chebwin)
windows = {'hamming'       : hamming(32),
           'hanning'       : hanning(32),
           'kaiser'        : kaiser(32,5),
           'nuttall'       : nuttall(32),
           'blackman'      : blackman(32),
           'blackmanharris': blackmanharris(32),
           'bartlett'      : bartlett(32)
           }
for key in windows.iterkeys():
    plot(windows.get(key), label=key)
axis([0, 31, 0, 1])
grid()
legend()
```

```python
figsize(10,10)
y = np.random.normal(0, 0.1, size=(512,))
plot(arange(len(y)), y, 'b')
for key in windows.iterkeys():
    window = windows[key]
    smoothed = convolve(y, window/window.sum())
    plot(arange(512), smoothed[:512], label=key)
legend()
```
